# Data Science with Python Pandas

## Exploring Series

Series
* One dimension array-like object 
* Capable of holding any data type
* Has an index

```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series(np.random.randn(5), index=['a','b','c','d','e']); print s
```

```python
print s[0]
```

## Exploring DataFrame

DataFrame
* Two dimensional tabular data structure
* Capable of holding any/many data types
* Index and columns

```python
df = pd.DataFrame(s, columns = ['column1']); df
```

```python
df['column1']
```

```python
df['column2'] = df['column1']*4; df
```

```python
df.sort_values(by='column1')
```

```python
# Boolean Indexing
df[df['column2'] < 2]
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
% matplotlib inline
plt.style.use('ggplot')
```

```python
df = pd.read_csv('GlobalTemperatures.csv')
```

```python
df.head()
```

```python
# considering only the 'LandAverageTempratuer' column
df = df.iloc[:, :2]
# df = df.ix[:, :2]
df.head()
```

```python
df.describe()
```

```python
type(df)
```

```python
type(df['LandAvereageTemperature'])
```
